```python
import tkinter as tk
import winsound  # built-in Windows sound module

root = tk.Tk()
root.title("Button Sounds")

# define function to play sound
def play_sound(filename):
    try:
        winsound.PlaySound(filename, winsound.SND_FILENAME | winsound.SND_ASYNC)
    except Exception as e:
        print(f"Error: {e}")

# create buttons and assign sound files to them
button1 = tk.Button(root, text="Button 1", command=lambda: play_sound('sound1.wav'))
button2 = tk.Button(root, text="Button 2", command=lambda: play_sound('sound2.wav'))

# pack the buttons on the window
button1.pack(pady=10)
button2.pack(pady=10)

root.mainloop()
```

```python
from tkinter import *
import tkinter as tk
from PIL import Image, ImageTk

def space(string):
  try:
    if " " in string:
      raise ValueError("Space is not valid")
    else:
      return True
  except ValueError as error:
    print("Error occured :",error)

def unary_minus(statement):
  try:
    if '=-' not in statement:
      return True
    else:
      raise ValueError("Unary minus is not valid")
  except ValueError as error:
    print("Error occured :",error)

def is_operator(op):
  if op == '+' or op == '-' or op == '*' or op == '%' or op == '/':
    return True
  else:
    return False


from io import StringIO
import tokenize

def exp_lst(exp):
    return [token[1] for token in tokenize.generate_tokens(StringIO(exp).readline) if token[1]]


def valid_expression(expression):
  c = 0
  exp=exp_lst(expression)
  for i in exp:
    if is_operator(i):
      c += 1
  try:
    if c>1:
      raise ValueError("Invalid Expression")
    else:
      return True
  except ValueError as error:
    print("Error occured :",error)



def is_variable(var):
  try:
    if var not in ['A','B','C','D','E']:
      raise ValueError("Invalid variable name")
    else:
      return True
  except ValueError as error:
    print("Error occured :",error)


# define the interpreter function
def interpreter(statements):
    memory = {}
    for statement in statements:
        if statement == 'STOP':
            break
        if unary_minus(statement) and space(statement):
            var, expression = statement.split('=')
            var = var.strip()
            if is_variable(var):
                # print(var,expression)
                expression = expression.strip()
                if valid_expression(expression):
                    if expression.isdigit() and int(expression) < 1000:
                        memory[var] = int(expression)
                    elif '+' in expression or '-' in expression or '*' in expression:
                        for i in range(len(expression)):
                            if expression[i] in ['+', '-', '*']:
                                operator_index = i
                                break
                        constituent1 = expression[:operator_index].strip()
                        operator = expression[operator_index].strip()
                        constituent2 = expression[operator_index + 1:].strip()
                        # print(constituent1,operator,constituent2)
                        if operator not in ['+', '-', '*']:
                            raise ValueError("Invalid operator")
                        if constituent1.isdigit() and int(constituent1) < 100:
                            constituent1 = int(constituent1)
                        elif constituent1 in memory:
                            constituent1 = memory[constituent1]
                        else:
                            raise ValueError("Invalid constituent")
                        if constituent2.isdigit() and int(constituent2) < 100:
                            constituent2 = int(constituent2)
                        elif constituent2 in memory:
                            constituent2 = memory[constituent2]
                        else:
                            raise ValueError("Invalid constituent")
                        if operator == '+':
                            result = constituent1 + constituent2
                        elif operator == '-':
                            result = constituent1 - constituent2
                        elif operator == '*':
                            result = constituent1 * constituent2
                        memory[var] = result
        else:
            raise ValueError("Invalid expression")
    sorted_dict = {key: value for key, value in sorted(memory.items())}
    result = ""
    for key, val in sorted_dict.items():
        result += key + ' = ' + str(val) + '\n'
    return result


# define GUI window for introduction
def intro_window():
    intro = tk.Tk()
    intro.title("Interpreter Program")
    intro.geometry("1600x1600")


    img = Image.open('C:\\Users\\hi\\Desktop\\WhatsApp Image 2023-03-05 at 7.07.16 PM.jpeg')
    bg = ImageTk.PhotoImage(img)
    lbl = Label(intro, image=bg)
    lbl.place(x=0, y=0)

    page1_button = tk.Button(intro, text="page1", font = ("bold", 15), height = 1, width = 8, command=input_window)
    page1_button.place(x = 900, y = 655)
    page1_button.configure(bg = 'cyan')

    page2_button = tk.Button(intro, text="page2", font = ("bold", 15), height = 1, width = 8, command=main)
    page2_button.place(x = 900, y = 705)
    page2_button.configure(bg = '#50C878')


    page3_button = tk.Button(intro, text='page3', font = ("bold", 15), height = 1, width = 8, command=ifelse)
    page3_button.place(x = 900, y = 755)
    page3_button.configure(bg = 'coral')

    intro.mainloop()


# define GUI window for input and output
def input_window():
    input = tk.Tk()
    input.title("Interpreter Program")
    input.geometry("1600x1600")
    input.configure(bg = '#E6C7C2')
    input_label = tk.Label(input, text="Enter your statements below:", font=("Arial", 30))
    input_label.pack(pady=10)
    input_text = tk.Text(input, height=7, font = ("Arial", 20))
    input_text.pack(padx=50, pady=10)

    execute_button = tk.Button(input, text="Execute", font=("bold", 20),
                               command=lambda: execute(input_text.get("1.0", "end-1c")))
    execute_button.pack(pady=10)

    output_label = tk.Label(input, text="Output:", font=("Arial", 30))
    output_label.pack(pady=10)
    output_text = tk.Text(input, height=7, font = ("Arial", 20))
    output_text.pack(padx=50, pady=10)

    button = Button(input, text="Close", font = ("bold", 20), command=lambda: close_iwindow())
    button.pack(side="bottom", padx=10, pady=35)

    def execute(statements):
        output_text.delete('1.0', tk.END)
        try:
            memory = interpreter(statements.split('\n'))
            if isinstance(memory, str):
                output_text.insert(tk.END, memory)
            else:
                sorted_dict = {key: value for key, value in sorted(memory.items())}
                for key, value in sorted_dict.items():
                    result = f"{key} = {value}\n"
                    output_text.insert(tk.END, result)
                output_text.insert(tk.END, "Execution successful.\n")
        except Exception as e:
            output_text.insert(tk.END, f"Error: {e}\n")

    def close_iwindow():
        input.destroy()

    input.mainloop()


def main():
    # Create the main window and widgets
    root = tk.Tk()
    root.title("Compiler GUI")
    root.geometry("1600x1600")
    root.configure(bg = "#FFCBA4")

    input_label = tk.Label(root, text="Input:", font = ("Arial", 30))
    input_label.pack(pady = 10)

    input_text = tk.Text(root, height=7, font = ("Arial", 20))
    input_text.pack(padx=50, pady=10)

    submit_button = tk.Button(root, text="Submit", font = ("bold", 20), command=lambda: print_argument())
    submit_button.pack(pady=10)

    output_label = tk.Label(root, text="Output:", font = ("Arial", 30))
    output_label.pack(pady = 10)

    output_text = tk.Text(root, height=7, font = ("Arial", 20))
    output_text.pack(padx=50, pady=10)

    button = Button(root, text="Close", font = ("bold", 20), command=lambda: close_main())
    button.pack(side="bottom", padx=10, pady=35)

    def print_argument():
        input_string = input_text.get("1.0", "end-1c")
        argument = input_string.split("'")[1].split("'")[0]
        output_text.insert(tk.END, argument)

    def close_main():
        root.destroy()

    root.mainloop()


# program for simple if-else conditions

def ifelse():
    root = tk.Tk()
    root.title("If-Else Condition Evaluator")
    root.geometry("1600x1600")
    root.configure(bg="#C8A2C8")

    input_frame = tk.Frame(root)
    input_frame.pack(side="top", fill="both", expand=True)

    input_label = tk.Label(input_frame, text="Input Conditions:", font = ("Arial", 30))
    input_label.pack(side="top", padx=10, pady=10)

    input_textbox = tk.Text(input_frame, height=6, font = ("Arial", 20))
    input_textbox.pack(side="top", padx=10, pady=10, fill="both", expand=True)

    button_frame = tk.Frame(root)
    button_frame.pack(side="top", fill="both", expand=True)

    evaluate_button = tk.Button(button_frame, text="Evaluate Condition", font = ("bold", 20), command=lambda: evaluate_conditions())
    evaluate_button.pack(side="left", padx=10, pady=10)

    evaluate_function_button = tk.Button(button_frame, text="Evaluate Function", font = ("bold", 20), command=lambda: evaluate_function())
    evaluate_function_button.pack(side="left", padx=10, pady=10)

    output_frame = tk.Frame(root)
    output_frame.pack(side="top", fill="both", expand=True)

    output_label = tk.Label(output_frame, text="Output:", font = ("Arial", 30))
    output_label.pack(side="top", padx=10, pady=10)

    output_textbox = tk.Text(output_frame, height=6, font = ("Arial", 20))
    output_textbox.pack(side="top", padx=10, pady=10, fill="both", expand=True)

    button = Button(root, text="Close", font = ("bold", 20), command=lambda: close_ifelse())
    button.pack(side="bottom", padx=10, pady=35)

    def evaluate_conditions():
        input_text = input_textbox.get("1.0", "end-1c")
        try:
            exec(input_text, globals())
            output_text = str(result) if 'result' in globals() else ""
        except Exception as e:
            output_text = "Error: " + str(e)
        output_textbox.delete("1.0", "end")
        output_textbox.insert("1.0", output_text)

    def evaluate_function():
        input_text = input_textbox.get("1.0", "end-1c")
        try:
            exec(input_text)
        except Exception as e:
            output_text = "Error: " + str(e)
        else:
            output_text = ""
            for name, obj in locals().items():
                if callable(obj):
                    output_text += f"{name}()\n"
                    output_text += "-" * 50 + "\n"
                    try:
                        output_text += str(obj()) + "\n"
                    except Exception as e:
                        output_text += "Error: " + str(e) + "\n"
                    output_text += "-" * 50 + "\n\n"
        output_textbox.delete("1.0", "end")
        output_textbox.insert("1.0", output_text)

    def close_ifelse():
        root.destroy()

    root.mainloop()

# start program by opening introduction window
intro_window()
```
